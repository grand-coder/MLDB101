```python
import pandas as pd
train_df = pd.read_csv("poem_train.csv", encoding="utf-8")
train_df
```

```python
test_df = pd.read_csv("poem_test.csv", encoding="utf-8")
test_df
```

```python
import jieba
s = "平林漠漠煙如織，寒山一帶傷心碧。\r\n暝色入高樓，有人樓上愁。"
" ".join(jieba.cut(s)).replace("\r", "").replace("\n", "")
```

```python
def poemcut(s):
    return " ".join(jieba.cut(s)).replace("\r", "").replace("\n", "")
# 先拿出一行(Series).apply(流程名字)
x_train = train_df["內容"].apply(poemcut)
x_test = test_df["內容"].apply(poemcut)
x_train
```

```python
# 準備一個字典   Series.replace(字典)
trans = {"李白":0, "杜甫":1, "白居易":2}
y_train = train_df["作者"].replace(trans)
y_test = test_df["作者"].replace(trans)
y_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
x_train_vec = vec.fit_transform(x_train)
x_test_vec = vec.transform(x_test)
```

```python
print(len(vec.vocabulary_))
vec.vocabulary_
```

```python
print(x_train_vec)
```

```python
reverse = {vec.vocabulary_[k]:k for k in vec.vocabulary_}
reverse[30000]
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(x_train_vec, y_train)
```

```python
from sklearn.metrics import accuracy_score
pre = clf.predict(x_test_vec)
print("預測:", list(pre))
print("正確:", list(y_test))
print(accuracy_score(pre, y_test))
```

```python
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(y_test, pre))
```
